# **Using Arabic Sentences**

```python
arabic_text = "قامت الشركه بأصدار بيانات تفصيليه حول ادائها فى الربع الأول من العام الحالى"
```

```python
#tokenize text into words
words = word_tokenize(arabic_text)
print(words)
print("length : ", len(words))
```

```python
#Load the Arabic Stop_Words
stop_words = set(stopwords.words('arabic'))
print(stop_words)
```

```python
#filter stopwords from text
filtered_words = [word for word in words if not word in stop_words]
print(filtered_words)
print("length : ", len(filtered_words))
```

```python
# Join the filtered words into a string
filtered_text = ' '.join(filtered_words)
print(filtered_text)
print("length : ", len(filtered_words))
```
