# **Word Tokenization**

```python
import nltk
nltk.download('punkt')
```

```python
text = "Hello! How are you doing today? I'm doing great, thanks for asking."
```

```python
# Tokenize Text Into Individual Sentences
sentences = nltk.sent_tokenize(text)
print(sentences)
```

```python
tokents = nltk.word_tokenize(text)
print(tokents)
```

# **Using txt->Dataset**

```python
import pandas as pd

# Read the entire file into a single string
with open("/content/lyrics-Alanwalker(8).txt", "r") as file:
    text_data = file.read()

# Create a DataFrame with a single column named 'lyrics'
dataa = pd.DataFrame({"lyrics": [text_data]})

# Now 'dataa' is a DataFrame with your lyrics in a single column
print(dataa)
```

```python
dataa.head()
```

```python
# Access the lyrics string from the DataFrame
lyrics_text = dataa['lyrics'].values[0]
print(lyrics_text)
```

```python
sentences = nltk.sent_tokenize(lyrics_text)

print(sentences)
```

```python
tokents = nltk.word_tokenize(lyrics_text)

print(tokents)
```
